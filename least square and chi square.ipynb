```python
#least square curve fitting

import numpy as np
import matplotlib.pyplot as plt
n=13
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]

#finding the coeffients
sumx=0
sumy=0
sumxy=0
sumx2=0

for i in range(n):
    sumx=sumx+x[i]
    sumy=sumy+y[i]
    sumx2=sumx2+(x[i]**2)
    sumxy=sumxy+(x[i]*y[i])
    
m=(n*sumxy-sumx*sumy)/(n*sumx2-(sumx)**2)
c=(sumy-m*sumx)/n

print("m=",m)
print("c=",c)

yc=np.zeros(n)
for i in range(n):
    yc[i]=m*x[i]+c

plt.plot(x,yc,'r',label="fitted line")
plt.plot(x,y,'ob',label="data points")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.savefig('linearfit.pdf', format='pdf')
plt.show()
```

```python
# chi square

import numpy as np
import matplotlib.pyplot as plt

no_of_observations = 13
z = np.array([0.07, 0.04, 0.12, 0.17, 0.1791, 0.1993, 0.2, 0.27, 0.28, 0.35, 0.3519, 0.3802, 0.4])
H_observed = np.array([69, 69, 68.6, 83, 75, 75, 72.9, 77, 88.8, 82.7, 83, 83, 95])
std_deviation = np.array([19.6, 12, 26.2, 8, 4, 5, 29.6, 14, 36.6, 8.4, 14, 13.5, 17])

H0 = 67.8

omega_m = np.linspace(0, 1, 1000)
chi_square_values = []

for i in range(len(omega_m)):
    chi_square = 0
    for j in range(no_of_observations):
        H_theor = H0 * (np.sqrt(omega_m[i] * (1 + z[j]) ** 3 + (1 - omega_m[i])))
        chi_square += ((H_observed[j] - H_theor) / std_deviation[j]) ** 2
    chi_square_values.append(chi_square)

minimum_chi_square = min(chi_square_values)
print("Minimum Chi-square:", minimum_chi_square)

index = chi_square_values.index(minimum_chi_square)
best_omega_m = omega_m[index]
print("Best Omega_M:", best_omega_m)
plt.plot(omega_m,chi_square_values)
plt.title('\u03A7\u00b2 v/s \u03A9m')
plt.xlabel('\u03A9m')
plt.ylabel('\u03A7\u00b2')
plt.grid()
plt.savefig('chi2.pdf', format='pdf')
plt.show()
```
